```python
# 변환
import json
import pandas as pd
import os


csv_file_path = "/content/pick.csv"
json_file_path = "/content/output_test.json"
output_file_path = "/content/filtered_output.json"


csv_data = pd.read_csv(csv_file_path)


with open(json_file_path, 'r', encoding='utf-8') as f:
    json_data = json.load(f)


file_names_to_include = csv_data['file_name'].tolist()  # raw_123 형식의 리스트


filtered_json_data = [
    item for item in json_data
    if os.path.splitext(item.get('file_name', '').lower())[0] in file_names_to_include
]

# 5. 새 ID 부여 (0부터 시작)
for idx, item in enumerate(filtered_json_data):
    item['id'] = idx

# 6. update json
with open(output_file_path, 'w', encoding='utf-8') as f:
    json.dump(filtered_json_data, f, ensure_ascii=False, indent=4)

# 7. 필터링된 데이터 개수 출력
print(f"필터링된 파일 개수: {len(filtered_json_data)}")
print(f"필터링된 JSON 데이터가 {output_file_path}에 저장되었습니다.")
```

```python
import json
import pandas as pd
import os

csv_file_path = "/content/pick.csv"
json_file_path = "/content/output_test.json"


csv_data = pd.read_csv(csv_file_path)

with open(json_file_path, 'r', encoding='utf-8') as f:
    json_data = json.load(f)

#  공백 제거 리스트 추출
file_names_to_include = [file_name.strip() for file_name in csv_data['file_name'].tolist()]

#  JSON에서 확장자 제거 jpg, JPG 이슈
json_file_names = [os.path.splitext(os.path.basename(item.get('file_name', '').lower()))[0] for item in json_data]

#  매칭 안된 파일
unmatched_files = [name for name in file_names_to_include if name not in json_file_names]

#  출력
print(f"CSV 파일 중 pick한 json에서 매칭되지 않은 파일 이름 ({len(unmatched_files)}건):")
print(unmatched_files)
```
